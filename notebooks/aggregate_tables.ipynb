```python
import pandas as pd

print("dữ liệu xử lý...")
df = pd.read_csv('../data/processed/superstore_final.csv')
```

# AGG_EXECUTIVE (Dành cho C-Level: KPI theo Năm/Tháng)

```python

agg_executive = df.groupby(['order_year', 'order_month']).agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum'),
    total_orders=('order_id', 'nunique')
).reset_index()
agg_executive['profit_margin'] = agg_executive['total_profit'] / agg_executive['total_sales']
agg_executive.to_csv('../data/processed/aggregates/agg_executive.csv', index=False)
```

# AGG_PRODUCT_LOSS (Dành cho CFO: Phân tích bẫy giảm giá)

```python
agg_product_loss = df.groupby(['sub_category', 'discount_level']).agg(
    total_sales=('sales', 'sum'),
    total_profit=('profit', 'sum'),
    order_count=('order_id', 'count'),
    loss_orders_count=('is_loss_maker', 'sum')
).reset_index()
agg_product_loss['loss_rate'] = agg_product_loss['loss_orders_count'] / agg_product_loss['order_count']
agg_product_loss.to_csv('../data/processed/aggregates/agg_product_loss.csv', index=False)
```

# AGG_RFM_SEGMENTS (Dành cho CMO: Phân khúc khách hàng)

```python
agg_rfm_segments = df.groupby('customer_segment').agg(
    customer_count=('customer_id', 'nunique'),
    avg_recency=('recency', 'mean'),
    avg_frequency=('frequency', 'mean'),
    avg_monetary=('monetary', 'mean'),
    total_segment_sales=('sales', 'sum')
).reset_index()
agg_rfm_segments.to_csv('../data/processed/aggregates/agg_rfm_segments.csv', index=False)
```
